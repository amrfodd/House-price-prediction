```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
pd.set_option('display.max_rows', 10)
# Show the full results
pd.set_option('display.max_column', None)
pd.set_option('display.max_rows', None)

matplotlib.rcParams['figure.figsize'] = (20, 10)
```

```python
df = pd.read_csv('../Data/Cleaned.csv')
df.head()
```

```python
df1 = df.copy()
```

```python
df1['price_per_sqft'] = df1['price'] * 100000 / df1['total_sqft']
```

```python
df1.head()
```

```python
df1.location.nunique()
```

```python
location = df1.groupby('location')['location'].agg('count')
location = location.sort_values(ascending = False)
location
```

```python
less_than_10 = location[location <= 10]
less_than_10
```

```python
df1.location = df1.location.apply(lambda x: 'other' if x in less_than_10.index else x)
```

```python
df1.to_csv('../Data/Engineered.csv', index = False)
```
